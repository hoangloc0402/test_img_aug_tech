```python
import torch
from torchvision import transforms
from efficientnet_pytorch import EfficientNet
from implementation.utils import *
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

### EfficientNet B0 w/ Random Erasing
https://arxiv.org/abs/1708.04896

```python
tsfm = [transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.3, scale=(0.02, 0.33), ratio=(0.3, 1.3), value=0, inplace=False)]

train_data, val_data, test_data = get_dataset(tsfm)

model = EfficientNet.from_name("efficientnet-b0", override_params={'num_classes': 101})
epochs = 30
batch_size = 128
```

```python
# GPU usage: 13533MiB 
train(model, train_data, val_data, epochs, batch_size)
```

```python
plot_graph()
```

```python
evaluate(model, test_data, print_accu=True)
```
