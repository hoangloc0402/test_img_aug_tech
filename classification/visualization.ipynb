```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

from implementation.cutout import Cutout
from implementation.autoaugment import AutoAugment
from implementation.randaugment import RandAugment
```

```python
ori_img = Image.open("images/elephant.jpg")

def show_comparision(img):
    NUM_ROWS = 1
    IMGs_IN_ROW = 2
    f, ax = plt.subplots(NUM_ROWS, IMGs_IN_ROW, figsize=(10,4))
    ax[0].imshow(ori_img)
    ax[1].imshow(img)
    ax[0].set_title('Before (original image)')
    ax[1].set_title('After')
    plt.tight_layout()
    plt.show()
```

## Random Erasing

```python
tfms = transforms.Compose([
                        transforms.ToTensor(),
                        transforms.RandomErasing(p=1, scale=(0.02, 0.33), ratio=(0.3, 1.3), value=0, inplace=False)])
img = transforms.functional.to_pil_image(tfms(ori_img))
show_comparision(img)
```

## ResizedCrop

```python
tfms = transforms.RandomResizedCrop((img.size[1], img.size[0]), scale=(0.5, 0.9), ratio=(0.75, 1.33))
img = tfms(ori_img)
show_comparision(img)
```

## Vertical Flip

```python
tfms = transforms.RandomVerticalFlip(p=1)
img = tfms(ori_img)
show_comparision(img)
```

## Horizontal Flip

```python
tfms = transforms.RandomHorizontalFlip(p=1)
img = tfms(ori_img)
show_comparision(img)
```

## Rotation

```python
tfms = transforms.RandomRotation(degrees=25)
img = tfms(ori_img)
show_comparision(img)
```

## Affine Transform

```python
tfms = transforms.RandomAffine(degrees=15, scale=(0.6, 0.9), shear=50)
img = tfms(ori_img)
show_comparision(img)
```

## Cutout

```python
tfms = transforms.Compose([
                    transforms.ToTensor(),
                    Cutout(n_holes=2, length=50)])
img = transforms.functional.to_pil_image(tfms(ori_img))
show_comparision(img)
```

## AutoAugment

```python
#AutoAugment will randomly give different result
tfms = AutoAugment()
img = tfms(ori_img)
show_comparision(img)
```

```python
tfms = AutoAugment()
img = tfms(ori_img)
show_comparision(img)
```

## RandAugment

```python
#RandAugment will randomly give different result
tfms = RandAugment()
img = tfms(ori_img)
show_comparision(img)
```

```python
tfms = RandAugment()
img = tfms(ori_img)
show_comparision(img)
```
