```python
import torch
from torchvision import transforms
from PIL import Image
```

## Original Image

```python
img = Image.open("/home/loc/projects/efficient_net/data/101_ObjectCategories/elephant/image_0001.jpg")
img
```

## Random Erasing

```python
a = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomErasing(p=1, scale=(0.02, 0.33), ratio=(0.3, 1.3), value=0, inplace=False)])(img)
transforms.functional.to_pil_image(a)
```

## ResizedCrop

```python
transforms.RandomResizedCrop((img.size[1], img.size[0]), scale=(0.5, 0.9), ratio=(0.75, 1.33))(img)
```

## Vertical Flip

```python
transforms.RandomVerticalFlip(p=1)(img)
```

## Horizontal Flip

```python
transforms.RandomHorizontalFlip(p=1)(img)
```

## Rotation

```python
transforms.RandomRotation(degrees=25)(img)
```

## Affine Transform

```python
transforms.RandomAffine(degrees=15, scale=(0.6, 0.9), shear=50)(img)
```
